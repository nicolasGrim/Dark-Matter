```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nu_e = pd.read_csv('data/effArea/nu_e.csv', delimiter = ';', names = ['E', 'Aeff'], decimal = ',')
nu_mu = pd.read_csv('data/effArea/nu_mu.csv',  delimiter = ';', names = ['E', 'Aeff'], decimal = ',')
nu_tau = pd.read_csv('data/effArea/nu_tau.csv',  delimiter = ';', names = ['E', 'Aeff'], decimal = ',')
```

```python
fig, ax = plt.subplots(1, 1, dpi = 300, figsize = (16, 9))

nu_e.plot(kind = 'scatter', x = 'E', y = 'Aeff', ax = ax, loglog = True, color = 'b', label = '$\\nu_e$')
nu_mu.plot(kind = 'scatter', x = 'E', y = 'Aeff', ax = ax, loglog = True, color = 'r', label = '$\\nu_\\mu$')
nu_tau.plot(kind = 'scatter', x = 'E', y = 'Aeff', ax = ax, loglog = True, color = 'g', label = '$\\nu_\\tau$')

ax.set_ylabel('Effective area [$m^2$]')
ax.set_xlabel('$E_{MC}$ [GeV]')

ax.legend()
ax.grid()

ax.set_xlim(10, 1000)
ax.set_ylim(1e-7, 1e-1)

fig.tight_layout()
```
